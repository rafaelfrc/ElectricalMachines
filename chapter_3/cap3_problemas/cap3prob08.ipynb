##### Variáveis:
 
 V_0: tensão da fonte [V]

```python
V_0, q, C, t, R = symbols('V_0 q C t R')
```

##### Tensão no capacitor [V]:

Se tomarmos um ponto qualquer do circuito após a abertura da chave, temos que, pela Lei dos nós de Kirchhoff:

$$i_R =  - i_C$$

$$\frac{v_C}{R} = - C\frac{dv_C}{dt}$$

$$\int \frac{1}{v_C} dv_C = - \int \frac{dt}{RC}$$

$$ \ln{v_C} = - \frac{t}{RC} + k$$

$$ e^{-\frac{t}{RC} + k} =  v_C$$

$$v_C = e^{k} e^{-\frac{t}{RC}}$$

$$v_C = V_0 e^{-\frac{t}{RC}}$$

```python
v_C = V_0*exp(-t/(R*C))
v_C
```

Carga no capacitor [C]:

```python
q = C*v_C
q
```

Energia armazenada [J]:

```python
W = (1/2)*(q**2)/C
W
```

Potência dissipada [W]:

```python
P = (v_C**2)/R
P
```

Energia total dissipada no resistor:

```python
Etotal = integrate(P, (t, 0, oo))
Etotal
```
